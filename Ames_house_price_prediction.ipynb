```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')
df = train.copy()
df.head()
```

```python
print('Shape of dataframe = ', df.shape)
print()
df.info()
```

```python
df.describe()
```

```python
cat_columns = [x for x in list(df.columns) if df[x].dtypes == 'object']
num_columns = [x for x in df.columns if x not in cat_columns]
print('No. of categorical features: ', len(cat_columns))
print('No. of numerical features: ', len(num_columns))
```

```python
fig, axs = plt.subplots(nrows=9, ncols=5, figsize=(30, 30))
for i, feat in enumerate(cat_columns):
    sns.countplot(data=df, x=feat, ax=axs[i % 9, i % 5])
```

```python
fig, axs = plt.subplots(nrows=8, ncols=5, figsize=(30, 30))
for i, feat in enumerate(num_columns):
    sns.histplot(data=df, x=feat, ax=axs[i % 8, i % 5], kde=True)
```

```python
q1 = df.GrLivArea.quantile(0.25)
q3 = df.GrLivArea.quantile(0.75)
iqr = q3-q1
df = df[(df.GrLivArea >= q1-1.5*iqr) & (df.GrLivArea <= q3+1.5*iqr)]
sns.histplot(data=df, x='GrLivArea', kde=True)
```

```python
cols = ['GrLivArea','BsmtFinSF1', 'LotFrontage', 'LotArea', 'GarageArea', 'TotalBsmtSF', '1stFlrSF', 'SalePrice']
for i in cols:
    q1 = df[i].quantile(0.25)
    q3 = df[i].quantile(0.75)
    iqr = q3-q1
    df = df[(df[i] >= q1-1.5*iqr) & (df[i] <= q3+1.5*iqr)]
    
fig, axs = plt.subplots(ncols=8, figsize=(20,5))
for i, feat in enumerate(cols):
    sns.histplot(data=df, x=feat, kde=True, ax= axs[i])
```

```python
null_val = df.isna().sum()
null_val[null_val > 0].sort_values(ascending=False)
```

```python
col_with_missing_vals = ['PoolQC', 'MiscFeature', 'Alley', 'Fence', 'FireplaceQu']
col_with_low_variance = ['GarageCond', 'Condition2', 'Street',
                         'Functional', 'Utilities', 'RoofMatl', 'Heating', 'GarageQual',
                         'LowQualFinSF', '3SsnPorch', 'KitchenAbvGr', 'MiscVal', 
                         'PoolArea', 'BsmtHalfBath']

drop_cols = col_with_missing_vals+col_with_low_variance
drop_cols.append('Id')
df.drop(drop_cols, axis=1, inplace=True)
df.shape
```

```python
df2 = df.copy()
corr_cols = ['OverallQual', 'YearBuilt', 'YearRemodAdd', 'GrLivArea','TotalBsmtSF','1stFlrSF', 
             'FullBath', 'TotRmsAbvGrd', 'GarageCars', 'GarageArea', 'SalePrice']
df2 = df2[corr_cols]
df2.head()
```

```python
corr_cols.remove('SalePrice')
fig,axs = plt.subplots(nrows=5, ncols=2, figsize=(15,15))
for i,feat in enumerate(corr_cols):
    sns.regplot(data=df2, x=feat, y='SalePrice',ax=axs[i%5, i%2], )
plt.tight_layout()
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import OneHotEncoder
from sklearn.decomposition import PCA
from sklearn.model_selection import cross_val_score
```

```python
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression(normalize=True)
X = df2.drop('SalePrice', axis=1)
y = df2.SalePrice
cv_score = cross_val_score(lin_reg, X, y, scoring='neg_mean_absolute_error', cv=5)
print("Average Cross Val Score = ", -cv_score.mean())
```

```python
cat_columns = [x for x in list(df.columns) if df[x].dtypes == 'object']
num_columns = [x for x in df.columns if x not in cat_columns]
num_columns.remove('SalePrice')
num_transformer = Pipeline(steps = [('Impute Num var' , SimpleImputer(strategy='mean')),
                                    ('Stadardization', StandardScaler())])
cat_transformer = Pipeline([('Impute Cat vars' , SimpleImputer(strategy='most_frequent')),
                            ('OneHot encoding', OneHotEncoder(sparse=False, handle_unknown='ignore'))])
t = [('Transform categorical Variables', cat_transformer, cat_columns),
     ('Transform numerical variables', num_transformer, num_columns)]
transformer = ColumnTransformer(transformers= t)
model = LinearRegression()
my_pipeline = Pipeline(steps= [('Preprocessing', transformer),
                               ('PCA', PCA(0.85)),
                               ('Model', model)])
```

```python
from sklearn.metrics import mean_absolute_error
X = df.drop('SalePrice', axis=1)
y = df.SalePrice
cv_score = cross_val_score(my_pipeline, X, y, scoring='neg_mean_absolute_error',cv=5)
print('Average cross val score = ', -cv_score.mean())
```
